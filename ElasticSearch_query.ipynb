```python
from elasticsearch import Elasticsearch
 
ES_HOST = {"host" : "localhost", "port" : 9200}
es = Elasticsearch(hosts = [ES_HOST])

res = es.search(body= {
    "query": {
        "match": {
            "text" : "Flyers"
        }
    }
})
res['hits']
```
